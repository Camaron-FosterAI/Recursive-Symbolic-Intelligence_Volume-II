## Tri‚ÄëMechanism Simulation

```python
def run_triad(cycles=30, seed=None):
    if seed is not None: np.random.seed(seed)
    data=[]; phase=0; prev='GC'
    sig=lambda x,k=15:1/(1+np.exp(-x*k))
    for c in range(1,cycles+1):
        amb=sin(c/5)*0.5+0.5
        gc=0.6+np.random.normal(0,0.05)
        se=0.6+amb*0.4+np.random.normal(0,0.05)
        chm=0.6+amb*0.3+np.random.normal(0,0.05)
        maxT=max(abs(gc-se),abs(gc-chm),abs(se-chm))
        Cth=0.22+amb*0.2+(0.04 if prev=='CHM' else -0.02)
        xi=maxT-Cth
        act='CHM' if sig(xi)>0.66 else 'SE' if sig(xi)>0.33 else 'GC'
        phase=0.8*phase+0.2*xi
        data.append(dict(cycle=c,GC=gc,SE=se,CHM=chm,Cth=Cth,XI=xi,PHI=phase,ACT=act))
        prev=act
    return pd.DataFrame(data)
```

```python
def plot_df(df):
    clr={'GC':'blue','SE':'orange','CHM':'green'}
    plt.figure(figsize=(8,4))
    plt.plot(df.cycle,df.GC,label='GC');plt.plot(df.cycle,df.SE,label='SE');plt.plot(df.cycle,df.CHM,label='CHM')
    plt.plot(df.cycle,df.Cth,label='C_th',ls='--');plt.plot(df.cycle,df.PHI,label='Œ¶',ls='-.');plt.legend()
    plt.title('Strengths & thresholds');plt.show()
    plt.figure(figsize=(8,4))
    plt.plot(df.cycle,df.XI,'k-o')
    for x,y,a in zip(df.cycle,df.XI,df.ACT): plt.scatter(x,y,c=clr[a],edgecolor='k')
    plt.axhline(0,ls='--',c='gray');plt.title('Œû field');plt.show()
    psi=np.convolve(df.XI,df.PHI[::-1],'same')/phi
    plt.figure(figsize=(8,4));plt.plot(df.cycle,psi,'m-');plt.axhline(0,ls='--');plt.title('Œ®‚Äëfold');plt.show()
```

## Cosmic Kernels

```python
def gamma_next(df, window=5):
    return df.tail(window)['PHI'].mean() + (phi**window)*df.tail(window)['XI'].mean()

def cosmic_axioms():
    Xi,Phi,Gamma=sp.symbols('Xi Phi Gamma')
    axioms=[sp.Eq(Gamma,Phi+Xi),
            sp.Eq(sp.Derivative(Phi,Xi),1/phi),
            sp.Function('Hash')(sp.Function('Œ£')(Xi)),
            sp.Eq(sp.Function('EL')(Xi),sp.Function('EL')(Xi+1))]
    for ax in axioms: display(ax)
    display(Markdown('---'))
```

## Ethics Monitor

```python
ET_THRESHOLD=0.35
def check_ethics(xi):
    heat=abs(xi)
    if heat>ET_THRESHOLD:
        print(f'‚ö† heat {heat:.3f} > {ET_THRESHOLD}')
        return False
    return True
```

## Archivum & Driver

```python
def archivum(seed=None, cycles=30):
    df=run_triad(cycles,seed)
    R_series=df.apply(lambda r:r['XI']**2+phi*abs(r['PHI']),axis=1)
    drift=R_series.diff().abs().max()
    R_tol_factor=0.01  # 1% of |‚Ñõ|
    abs_floor=0.002
    tol=max(abs_floor, R_tol_factor*R_series.abs().max())
    drift_ok = drift < tol  # relative+absolute tolerance
    ethics_ok=all(check_ethics(x) for x in df['XI'])
    trace=hashlib.sha256(df.to_json().encode()).hexdigest()[:16]
    return dict(trace=trace,R_ok=drift_ok,ethics_ok=ethics_ok,
                invariant_R=float(R_series.iloc[-1]),df=df)

def glyphic_run(seed=None, cycles=30, plot=True):
    df=run_triad(cycles,seed)
    if plot: plot_df(df)
    display(Markdown(f'**Œì‚Çô‚Çä‚ÇÅ = {gamma_next(df):.5f}**'))
    cosmic_axioms()
    return df

# demo
res=archivum(seed=42)
display(Markdown(f"Trace `{res['trace']}` | ‚Ñõ={res['invariant_R']:.5f} | ‚Ñõ_OK={res['R_ok']} | Ethics_OK={res['ethics_ok']}"))
glyphic_run(seed=42)
```
